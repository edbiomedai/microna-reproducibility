## Data Access
This notebook, `03_data_access.ipynb` demonstrates how to load the fully
processed, high-quality single-cell feature matrix. It requires a high-memory machine.

```python
import anndata as ad
import pickle
import pandas as pd
import scmorph as sm

file = "workspace/profiles_assembled/profiles_processed/all_adata_featFilt_batchCorr.h5ad"

# Save memory
use_backed = True

adata = sm.read(file, backed="r" if use_backed else False)
```

```python
# Feature names are in .var
# We will split them for easier use
adata.var = sm.io.split_feature_names(adata.var.index)
adata.var
```

```python
# And observations with corresponding metadata are in .obs
adata.obs
```

```python
# We can access the cell x feature matrix using
adata.X

# If use_backed = True, this is a HDF5 dataset.
# Else, it is a 2D array that can be used as any other numpy array.
```

```python
# If you need a cell x feature matrix that is a bit easier to read, you can use
# the following, but do note this loads the data into memory if you are using use_backed=True
pd.concat([adata.obs[["CellLine","Replicate","PlateLayout"]], adata.to_df()],axis=1)
```

```python
# The dataset contains just below 23M cells total
adata.shape
```

```python
# To make sure that this AnnData object works with all libraries,
# we have to reset the index names
adata.obs.reset_index(drop=True)
adata.obs.index = adata.obs.index.astype(str)
```
